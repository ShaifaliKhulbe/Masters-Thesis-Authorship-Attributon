```python
import pandas as pd
import nltk

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('Test_Hindi_data.csv')
#df = df.iloc[:1000, :]

# Create a new DataFrame with columns "Author_name", "chunks", and "chunk_length"
new_df = pd.DataFrame(columns=["Author_name", "chunks", "chunk_length"])

# Group the rows by the "Author_name" column
groups = df.groupby("Author_name")

# Iterate through each group of rows
for name, group in groups:

    # Create a list of sentences in the current group
    sentences = list(group["Sentences"])

    # Create a list to hold the chunks for the current author
    chunks = ""

    # Create a variable to keep track of the current chunk's length
    current_length = 0

    # Iterate through each sentence in the text
    for sentence in sentences:

        # Tokenize the sentence using nltk
        tokens = nltk.word_tokenize(sentence)

        # Get the length of the sentence in terms of tokens
        sentence_length = len(tokens)
        
        # If adding the sentence to the current chunk would result in a chunk longer than 200 tokens,
        # add the current chunk to the list of chunks and start a new chunk
        if current_length + sentence_length >= 200:
            current_length += sentence_length
            chunks = chunks + " " + sentence
            #new_chunk = " ".join(chunks)
            new_df = new_df.append({"Author_name": name, "chunks": chunks, "chunk_length": current_length}, ignore_index=True)
            chunks = ""
            current_length = 0
            sentence = ""
            sentence_length = 0

        # Add the sentence to the current chunk and update the current length
        chunks = chunks + " " + sentence
        current_length += sentence_length
        
    # If there's a remaining chunk at the end, add it to the new DataFrame row
    if chunks:
        #new_chunk = " ".join(chunks)
        new_df = new_df.append({"Author_name": name, "chunks": chunks, "chunk_length": current_length}, ignore_index=True)

```
